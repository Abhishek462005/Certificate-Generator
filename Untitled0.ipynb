```python
def player(prev_play, opponent_history=[], play_order={}):
    # Reset history for a new match
    if not prev_play:
        opponent_history.clear()

    opponent_history.append(prev_play)
    prediction = "R" # Default starting move

    # We need a certain amount of data before we can start predicting
    n = 3

    if len(opponent_history) > n:
        # Create a string of the last n moves
        last_n = "".join(opponent_history[-n:])

        # Track what move followed this specific sequence in the past
        # We look at the sequence ending just before the current move
        prev_sequence = "".join(opponent_history[-(n+1):-1])

        if prev_sequence:
            play_order[prev_sequence] = play_order.get(prev_sequence, {})
            play_order[prev_sequence][prev_play] = play_order[prev_sequence].get(prev_play, 0) + 1

            # Predict the next move based on the most frequent follow-up to the current sequence
            if last_n in play_order:
                potential_next_moves = play_order[last_n]
                prediction = max(potential_next_moves, key=potential_next_moves.get)

    # Counter the prediction
    ideal_response = {'P': 'S', 'R': 'P', 'S': 'R'}
    return ideal_response[prediction]
```
